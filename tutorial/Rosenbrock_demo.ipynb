```python
import optimizer.gradient_free 
import objective.test_optimizer 
import matplotlib.pyplot as plt 
import numpy as np 

# test Rosenbrock function: 1 global minimum ( 1,1 ) 


Rosenbrock_obj = objective.test_optimizer.Rosenbrock()  

optimize_Rosenbrock = optimizer.gradient_free.NelderMeadSimplex("in_rosenbrock","rosenbrock.log",Rosenbrock_obj,skipped=0)

optimize_Rosenbrock.run_optimization() 


x_limit = np.arange(-8,8,0.15) 

y_limit = np.arange(-8,8,0.15) 

Rosenbrock_test = objective.test_optimizer.Rosenbrock() 

z_mat = np.zeros((x_limit.size,y_limit.size))

counter_x = 0 

for x in x_limit: 

    counter_y = 0 

    for y in y_limit: 

        z_mat[counter_x,counter_y] = np.log10(Rosenbrock_test.evaluate(np.array([x,y])))  
    
        counter_y += 1 

    counter_x += 1 

for i in range(0,60,2):  

    vertices = np.loadtxt("vertices_%d.txt"%i)
        
    plt.scatter(vertices[:,0],vertices[:,1],s=1,color="r") 

    plt.plot(vertices[:4,0],vertices[:4,1],color="r") 

plt.contour(x_limit,y_limit,np.transpose(z_mat),20) 
plt.xlabel("x")
plt.ylabel("y")

plt.savefig("Rosenbrock_plot.png",dpi=300)

plt.show()

```
